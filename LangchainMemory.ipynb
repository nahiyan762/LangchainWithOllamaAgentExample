```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.messages import AIMessage, HumanMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_ollama import OllamaLLM
from langsmith import uuid7
```

```python
store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Remember the conversation context."),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

llm = OllamaLLM(model="deepseek-r1:1.5b", temperature=0.7)

chain = prompt | llm
```

```python
chain_with_history = RunnableWithMessageHistory(
    chain,
    get_session_history,
    input_messages_key="input",
    history_messages_key="history"
)
```

```python
id = uuid7()
config = {"configurable": {"session_id": id}}
```

```python
response1 = chain_with_history.invoke(
    {"input": "My name is Alex and I'm learning LangChain"},
    config=config
)
print("Response 1:", response1)

response2 = chain_with_history.invoke(
    {"input": "What's my name?"},
    config=config
)
print("Response 2:", response2)
```

```python
while True:
    print("=" *50)
    user_input = input("Enter your message: ")
    
    if user_input.lower() in ["exit", "quit"]:
        print("Exiting the chat. Goodbye!")
        break


    response = chain_with_history.invoke(
        {"input": user_input},
        config = config
    )

    print(f"AI: {response}")
```

```python
for messages in store[id]:
    for message in messages[1]:
        if isinstance (message, AIMessage):
            message_type = "AI"
        else:
            message_type = "HUMAN"
        print(f"{message.type}: {message}")
```
