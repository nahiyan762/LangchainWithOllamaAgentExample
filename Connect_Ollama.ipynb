```python
from langchain_core.prompts import PromptTemplate
from langchain_ollama import OllamaLLM
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate(
    template="Write a poem about {topic}",
    input_variables=["topic"]
)

llm = OllamaLLM(model="deepseek-r1:1.5b")

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "rain"}))
```

```python
from langchain_ollama import OllamaLLM
from langchain_community.tools import Tool
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

import math

llm = OllamaLLM(model="deepseek-r1:1.5b")

# Define a math tool manually
def calc(expression: str) -> str:
    return str(eval(expression))

math_tool = Tool(
    name="Calculator",
    func=calc,
    description="Useful for answering math expressions like '123 * 99'"
)

# Prompt for tool use
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI agent. Use tools when necessary."),
    ("human", "{input}")
])

# Build a simple agent
def agent_run(query):
    if any(op in query for op in ["+", "-", "*", "/", "**"]):
        expr = query.replace("What is", "").replace("?", "").strip()
        return math_tool.run(expr)
    else:
        chain = prompt | llm | StrOutputParser()
        return chain.invoke({"input": query})

print(agent_run("What is 123 * 99?"))
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.llms import Ollama

# Create a prompt template (ChatPromptTemplate produces a Runnable)
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful coding assistant specialized in {language}."),
    ("human", "Explain {concept} in {language} with a simple example.")
])

# Initialize model
llm = Ollama(model="deepseek-r1:1.5b")

# Compose a runnable chain via the | operator (no Chain class import needed)
chain = prompt | llm 

# Test the chain
result = chain.invoke({
    "language": "Python",
    "concept": "list comprehensions"
})

print(result)
```
